```python
repo_tree = { "TUTORIAL:SPOT-THE-MASK-CHALLENGE" :
        {'DATASET': [ 'CSV','DOWNLOAD', 'IMAGES', 'SHP','ZIP',] ,
        'NOTEBOOKS' : [ "CACHE" ] ,
        'SUBMISSIONS' : [ ] ,}
        }

```

```python

def create_repo_structure(base ='', tree=dict()):
    
    repo_tree = list(tree.values())[0]
    repo_name = list(tree.keys())[0]
    if base == '' :
        
        repo_path = os.path.join(os.getcwd(), repo_name)
    else:
        repo_path = os.path.join(base, repo_name)
    
    for key in repo_tree.keys():
        if  len(repo_tree[key]) == 0:
            folder_to_create = os.path.join(repo_path, key, )
            os.makedirs(folder_to_create, exist_ok=True)
            print(f"Created : '{os.path.relpath(folder_to_create)}' ")
        else:
            for value in repo_tree[key]:
                
                if isinstance( value,dict ) :
                    folder_to_create = os.path.join(repo_path, key, )
                    create_repo_structure(base =folder_to_create, tree=value)
                else:
                    folder_to_create = os.path.join(repo_path, key,value )
                    os.makedirs(folder_to_create, exist_ok=True)
                    print(f"Created : '{os.path.relpath(folder_to_create)}' ")
```

```python
create_repo_structure(tree=repo_tree)
```
